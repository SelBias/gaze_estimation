```python
import numpy as np
import random
from tqdm.auto import tqdm

np.random.seed(42)
indice_list = np.stack([np.random.choice(np.arange(160*144*2), 3000, replace=False) for _ in range(50)])
```

```python
converted_indice_list = [[] for _ in range(50)]
for i in tqdm(range(50)) :
    subj_ind = str(i).zfill(2)
    for j in range(3000) :
        eye_ind = f'left' if indice_list[i,j] % 2 == 0 else 'right'
        image_ind = (indice_list[i,j] // 2 % 144).astype(str).zfill(8)
        folder_ind = (indice_list[i,j] // 2 // 144).astype(str).zfill(3)

        ind = f's{subj_ind}/{folder_ind}_{eye_ind}/{image_ind}.bmp'

        converted_indice_list[i].append(ind)
ut_conv_indices = np.stack(converted_indice_list)
```

```python
np.save('./ut_sub_indices', indice_list)
np.save('./ut_sub_indices_str', ut_conv_indices)
```
